```python
def fetch_movie_data(page_number):
    """Fetches movie data from the API for a given page number."""
    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={page_number}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Request failed for page {page_number} with status code: {response.status_code}")
        return None

# Fetch data from multiple pages
all_movies = []
for page_number in range(1, 472):  # Change the range to scrape more pages
    data = fetch_movie_data(page_number)
    if data:
        for movie in data['results']:
            # Extract specific data for each movie
            movie_info = {
                'title': movie['title'],
                'overview': movie['overview'],
                'genre_ids': movie['genre_ids'],

            }
            all_movies.append(movie_info)

# Create a pandas DataFrame with selected columns
df = pd.DataFrame(all_movies)
print(df.head())
```

```python
# Remove HTML tags
import re
def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'',text)
```

```python
# Remove URL
def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'',text)
```

```python
# Chat word treatment
def chat_word_treatment(text):
  new_text = []
  for w in text.split():
    if w.upper() in chat_words:
      new_text.append(chat_words[w.upper()])
    else:
      new_text.append(w)
  return " ".join(new_text)
```

```python
df['overview'].apply(chat_word_treatment)
```

```python
# Spelling correction
from textblob import TextBlob
df['overview'] = df['overview'].apply(lambda x: str(TextBlob(x).correct()))
```

```python
# Removing stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

new_text = []
stop_words = stopwords.words('english')
def remove_stop_words(text):
  for word in text.split():
    if word in stop_words:
      new_text.append('')
    else:
      new_text.append(word)
  x = new_text[:]
  new_text.clear()
  return " ".join(x)
```

```python
df['overview'].apply(remove_stop_words)
```

```python
# Tokenize
import spacy
nlp = spacy.load('en_core_web_sm')
def tokenize(text):
  doc = nlp(text)
  return [token.text for token in doc]
```

```python
# stem words
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
  return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

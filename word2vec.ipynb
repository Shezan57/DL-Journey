```python
# import gensim and other necessary library
import pandas as pd
import numpy as np
import gensim
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
from zipfile import ZipFile
import os
```

```python
# unzip the dataset
with ZipFile('/content/001ssb.txt.zip','r') as Zip:
  os.makedirs('content/got_dataset',exist_ok=True)
  Zip.extractall(path='/content/got_dataset')
Zip.close()
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
from gensim.parsing.preprocessing import remove_stopwords
from nltk import sent_tokenize
```

```python
# remove stopwords and append to list
story = []
for filename in os.listdir('/content/got_dataset'):
  with open(os.path.join('/content/got_dataset',filename)) as f:
    files = f.read()
    stop_files = remove_stopwords(files)
    raw_sent = sent_tokenize(stop_files)
    for sent in raw_sent:
      story.append(simple_preprocess(sent))
```

```python
print(len(story))
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4
)
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar('daenerys')
```

```python
model.wv.doesnt_match(['jon','rikon','robb','arya','sansa','bran'])
```

```python
model.wv['king']
```
